```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
elonmusk_df = pd.read_csv("../data/raw/elonmusk.csv")
elonmusk_df.head()
```

```python
# Datetime Feature Extraction

elonmusk_df["Datetime"] = pd.to_datetime(elonmusk_df["Datetime"], errors="coerce")

elonmusk_df["Year"] = elonmusk_df["Datetime"].dt.year
elonmusk_df["Month"] = elonmusk_df["Datetime"].dt.month
elonmusk_df["Day"] = elonmusk_df["Datetime"].dt.day
elonmusk_df["Day of Week"] = elonmusk_df["date_column"].dt.dayofweek
```

```python
# Datetime Feature Preprocessing

elonmusk_df["Year"] = StandardScaler().fit_transform(elonmusk_df["Year"])

# Month, Day, and Day of Week are cyclic features so we encode using their natural cycles.
elonmusk_df["Month"] = np.sin(2 * np.pi * elonmusk_df["Month"] / 12)
elonmusk_df["Day"] = np.sin(2 * np.pi * elonmusk_df["Day"] / 31)
elonmusk_df["Day of Week"] = np.sin(2 * np.pi * elonmusk_df["Day of Week"] / 7)
```
