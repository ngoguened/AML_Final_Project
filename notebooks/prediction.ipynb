```python
import re

import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
elonmusk = pd.read_csv("../data/processed/elonmusk_processed.csv")
teslastock = pd.read_csv("../data/processed/teslastock_processed.csv")
```

```python
df = pd.merge(teslastock, elonmusk, on="Date", how="left")
print(df.columns)
```

```python
# Let's start first by seeing how good a linear regressor can predict the stock price with whether elon's tweet contains the word 'tesla'.

# First let's remove all non-overlapping dates in the dataset:
df = df[df["Day"].notna()]

df["Contains Tesla"] = (
    df["fullText"].str.contains("TESLA", flags=re.IGNORECASE, regex=True).astype(int)
)

X, y = df[["Year", "Month", "Day", "Contains Tesla"]], df["PriceChange"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)  # Not very good so far
```
