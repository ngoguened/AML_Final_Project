```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
teslastock_df = pd.read_csv("../data/raw/teslastock_raw.csv")
teslastock_df.head()
```

```python
# Date Feature Extraction

teslastock_df["Date"] = pd.to_datetime(teslastock_df["Date"], errors="coerce")
```

```python
# Price Feature Engineering
teslastock_df["PriceChange"] = teslastock_df["High"] - teslastock_df["Low"]
```

```python
teslastock_df.to_csv("../data/processed/teslastock_processed.csv")
```
